```python
import matplotlib.pyplot as plt
```

```python
n_qubits = [4, 8, 10]
tree_depths = [32, 65, 78]
tree_ecr_counts = [12, 34, 42]
line_depths = [41, 100, 124]
line_ecr_counts = [12, 34, 44]
```

```python
plt.plot(n_qubits, tree_depths, marker="o", label="Tree")
plt.plot(n_qubits, line_depths, marker="x", label="Line")
plt.xlabel("Number of qubits")
plt.ylabel("Depth")
plt.title("Depth vs Number of qubits on ibm_brisbane")
plt.legend()
plt.show()
```

```python
plt.plot(n_qubits, tree_ecr_counts, marker="o", label="Tree")
plt.plot(n_qubits, line_ecr_counts, marker="x", label="Line")
plt.xlabel("Number of qubits")
plt.ylabel("ECR count")
plt.title("ECR count vs Number of qubits on ibm_brisbane")
plt.legend()
plt.show()
```
