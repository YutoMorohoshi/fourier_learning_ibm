```python
import matplotlib.pyplot as plt
```

```python
n_qubits = [4, 8, 10]
tree_depths = [28, 62, 78]
tree_ecr_counts = [9, 27, 33]
line_depths = [36, 101, 124]
line_ecr_counts = [9, 27, 35]

qpu_backend = "ibm_torino"
```

```python
plt.plot(n_qubits, tree_depths, marker="o", label="Tree")
plt.plot(n_qubits, line_depths, marker="x", label="Line")
plt.xlabel("Number of qubits")
plt.ylabel("Depth")
plt.ylim(0, 150)
plt.title(f"Depth vs Number of qubits on {qpu_backend}")
plt.legend()
plt.show()
```

```python
plt.plot(n_qubits, tree_ecr_counts, marker="o", label="Tree")
plt.plot(n_qubits, line_ecr_counts, marker="x", label="Line")
plt.xlabel("Number of qubits")
plt.ylabel("ECR count")
plt.title(f"ECR count vs Number of qubits on {qpu_backend}")
plt.legend()
plt.show()
```
